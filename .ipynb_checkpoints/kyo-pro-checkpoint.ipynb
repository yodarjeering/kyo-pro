# Yokan Party
問題文
左右の長さが L [cm] のようかんがあります。 N 個の切れ目が付けられており、左から i 番目の切れ目は左から A 
i
​
  [cm] の位置にあります。

あなたは N 個の切れ目のうち K 個を選び、ようかんを K+1 個のピースに分割したいです。そこで、以下の値を スコア とします。

K+1 個のピースのうち、最も短いものの長さ（cm 単位）
スコアが最大となるように分割する場合に得られるスコアを求めてください。

制約
1≤K≤N≤100000
0<A 
1
​
 <A 
2
​
 <⋯<A 
N
​
 <L≤10 
9
 
入力はすべて整数

```python
n,l = map(int,input().split())
k = int(input())
a = list(map(int,input().split()))

def is_ok(a,m,k,l):
    cut_point = 0
    cut_count = 0
    
    for i in range(len(a)):
        delta = a[i] - cut_point
        if delta >= m:
            cut_point = a[i]
            cut_count += 1
            # 右の論理式は, 左だけ見てcutした羊羹を, 右からも見るための式
        if k<=cut_count and l-cut_point>=m:
            return True
    return False
    
right = l
left = 0
ans = 0
m = (right + left)//2
diff = right - left

while diff>1:

    
    print("-----------")
    print("m",m)
    print("right",right)
    print("left",left)
    if is_ok(a,m,k,l):
        ans = m
        left = m
        
    else:
        right = m
    m = (right + left)//2
    diff = right - left
print(m)
```

# 射撃王

高橋君は最近、射撃にハマっている。

高橋君は N 個の風船すべてを射撃で割り、得られる得点をできるだけ小さくする競技に参加している。

風船には 1 から N までの番号が付けられていて、風船 i(1≦i≦N) は競技開始時に高度 H 
i
​
  のところにあり、1 秒経過するにつれて高度が S 
i
​
  だけ増加する。

高橋君は競技開始時に 1 個風船を割ることができ、そこから 1 秒ごとに 1 個の風船を割ることができる。どの順番で風船を割るのかは高橋君が自由に決定できる。

どの風船についても、その風船を割ることによるペナルティが発生する。ペナルティはその風船が割られたときの高度と等しい整数値となる。高橋君が最終的に得る得点は N 個の風船のペナルティのうちの最大値となる。

高橋君が得ることのできる得点として考えられる最小値を求めよ。

https://www.slideshare.net/chokudai/abc023

```python
def BinarySearch(x, N, H, S):
    LimitTime = []
    for i in range(N):
        if x < H[i]:
            return False
#         Pi = Hi + ti*Si
# 上式の式変形から, ti = (x - H[i]) // S[i]
        LimitTime.append((x - H[i]) // S[i])
    LimitTime.sort()
    for i in range(N):
#         =　がついていないので, 重複を一個まで許すということか
#  重複, ti = 0の時はなんでもセーフだから
# 　この条件式は, 重複を2個以上許すとき, 同じ時間に風船を2個以上割ってしまうため, 組合せ的にありえないから
        if LimitTime[i] < i:
            return False
    return True

N = int(input())
L = [list(map(int,input().split())) for _ in range(N)]
H,S = [list(i) for i in zip(*L)]
ok = 10**18
ng = 0
while ok - ng > 1:
    x = (ok + ng) // 2
    if BinarySearch(x, N, H, S):
        ok = x
    else:
        ng = x
print(ok)
```

# 002 - Encyclopedia of Parentheses
問題文
長さ N の正しいカッコ列をすべて、辞書順に出力してください。

ただし、正しいカッコ列は次のように定義されています :

() は正しいカッコ列である
S が正しいカッコ列であるとき、文字列 ( +S+ ) は正しいカッコ列である
S,T が正しいカッコ列であるとき、文字列 S+T は正しいカッコ列である
それ以外の文字列はすべて、正しいカッコ列でない
例えば、

()()
(()())(())
()()()()()()()()
は正しいカッコ列ですが、

)(
)))()(((
((((a))))
は正しいカッコ列ではありません。

また、 ( の方が ) よりも辞書順で早いものとします。

見ないで解けた！

```python
n = int(input())
if n%2!=0:
    print()
    exit()

def binarize(x,keta):
    return bin(x).replace('0b','').zfill(keta)
  
def is_ok(bit):
    count1=0
    count0=0
    for i in range(len(bit)):
        if bit[i]=='1':
            count1+=1
        if bit[i]=='0':
            count0+=1
        if count0>count1:
            return False
    return True
  
keta = len(bin(2**n-1))-2
for i in reversed(range(2**n)):
    bit = binarize(i,keta)
    m = len(bit)//2
    if bit.count('1')==bit.count('0') and bit[0]!='0' and bit[-1]!='1' and is_ok(bit):
        ans = bit.replace('1','(').replace('0',')')
        print(ans)
          
  
```

入力はすべて整数

<h>最初に出した解答<h> 
これだとTLE

```python
n, b, k = map(int,input().split())
c = list(map(str,input().split()))

q_min = 10**(n-1) // b
q_max = 10**n // b
cnt = 0
r = 10**9 + 7

def is_ok(multiple_b):
    str_b = set(str(multiple_b))
    for b_ in str_b:
        if b_ not in c:
            return False
    return True

for i in range(q_min,q_max+1):
    multiple_b = b*i
    if is_ok(multiple_b):
    #print('multi_b',multiple_b)
        cnt = (cnt +1)%r

print(cnt)
    
```

むずすぎるため, 小課題のみクリアさせ, 倍数の個数が, こうなるのは暗記として理解する

```python
n, b, k_ = map(int,input().split())
c = list(map(int,input().split()))



r = 10**9 + 7
# 0桁はないから,　n桁目まで
dp = [[0 for col in range(b)] for row in range(n+1)]


for c_ in c:
    dp[1][c_%b] += 1
  
for i in range(n):
    for j in range(b):
        for k in c:
            nex = (10*j + k)%b
            dp[i+1][nex] += dp[i][j]
            dp[i+1][nex] %= r
      
print(dp[n][0])
    
```

最終的に答えた解答

```python
N,K = map(int,input().split())
S = input()

alphabet_dict = [[-1] * 26 for _ in range(N) ]


# ******** 前処理 *********
for i in reversed(range(N)):
    s = S[i]
#     後ろから ,　ある文字 i より右にある最左 (アルファベット順で若い) のアルファベット (ord(s) - ord('a')) のインデックスを代入して, 前処理していく
    
    
    if i<N-1:    
        for j in range(26):
            alphabet_dict[i][j] = alphabet_dict[i+1][j]
        
    alphabet_dict[i][ord(s)-ord('a')] = i
# ********* 前処理ここまで ********

index = 0
answer = []
for i in range(K):
    s = S[i]
    
    for j in range(26):
        next_index = alphabet_dict[index][j]
        
        if next_index==-1:
            continue
        
#         探索範囲を絞る
#         拘束条件  
        if N - K >= next_index - i:
            print("index",index)
            print("answer",answer)
            answer.append(chr(j + ord('a')))
            index = next_index+1
            break

            
print("".join(answer))
        
    
```

  です。 各生徒は自分に合わないレベルのクラスに行くと不満になります。 各生徒について、その不満度は次のように定義されます。

対象レーティング a と自分のレーティング b の差の絶対値、すなわち ∣a−b∣
j=1,2,3,…,Q それぞれについて、番号 j の生徒の不満度として考えられる最小値を求めてください。 ただし、1 人も入らないクラス、複数人から成るクラスがあっても良いものとします。

問題文
ABC 君にお仕事の依頼が N 件来ました。各仕事には 1,2,⋯,N と番号が付けられています。

仕事 i は締切が D 
i
​
  日目の終わりであり、連続する C 
i
​
  日間を使うことで完了できます。それ以外の方法で仕事を完了することはできません。

より正確には、仕事 i を締切までに完了するためには、C 
i
​
 ≤k≤D 
i
​
  を満たすある整数 k について k−C 
i
​
 +1 日目の始めから k 日目の終わりまで i 番目の仕事を続けなければなりません。

仕事 i を締切までに完了すると S 
i
​
  円の報酬がもらえます。1 日には高々 1 種類しか仕事を行うことができません。

現在は 1 日目の始まりです。ABC 君が上手く取り掛かる仕事を選んでスケジュールを組んだ場合、彼が得られる報酬の金額として考えられる最大値を求めてください。

```python
from operator import itemgetter

N = int(input())
D = []
C = []
S = []

# N = 4
# D = [7,5,3,10]
# C = [4,2,2,3]
# S = [2,3,4,2]

for i in range(N):
    d,c,s = map(int,input().split())
    D.append(d)
    C.append(c)
    S.append(s)
    
c = list(zip(D, C, S))
# Dでソート
c.sort(key=itemgetter(0))
D,C,S = zip(*c)

dp = [[0 for col in range(D[-1]*2)] for row in range(N+1)]

# dp のインデックスと, C,D,S のインデックスずれるのは注意

for i in range(N):
    for j in range(D[-1]):
#         仕事 i+1 をしない時
        dp[i+1][j] = max(dp[i+1][j],dp[i][j])
        
#         仕事 i+1 をするとき
        if j+C[i]<=D[i]:
            dp[i+1][j+C[i]] = max(dp[i+1][j+C[i]],dp[i][j]+S[i])
            
print(max(max(dp)))
```

```python
class UnionFind():
    
    
    def __init__(self, sz):
        self.par = [i for i in range(sz)]
        
        
    def root(self, pos):
        if self.par[pos]==pos:
            return pos
        
        self.par[pos] = self.root(self.par[pos])
        return self.par[pos]
    
    
    def unite(self,u,v):
        u = self.root(u)
        v = self.root(v)
        
        if u==v:
            return 
        
        self.par[u] = v
    
    
    def same(self,u,v):
        if self.root(u) == self.root(v):
            return True
        else:
            return False


H, W = map(int,input().split())
Q = int(input())


UF = UnionFind(H*W)
# 0 index 化していないので,　そのままの値でアクセスできるようにした
# W+2 としているのは　sx = px + 1 の時, out of range だから
# hash化した後のインデックスは0 index のためこの実装でよい
mass = [[0 for col in range(W+2)] for row in range(H+2)]


def query1(px,py):
    dx = [-1, 0, 1, 0]
    dy = [0, 1, 0, -1]
    
    for i in range(4):
        sx = px + dx[i]
        sy = py + dy[i]
        
        if mass[sx][sy] == 0:
            continue
            
#         隣接点が見つかった時
        hash1 = (px-1)*W + (py-1)
        hash2 = (sx-1)*W + (sy-1)
        UF.unite(hash1,hash2)
    mass[px][py] = 1
    

def query2(px,py,qx,qy):
#     多分計算速度上げるためのテク
    if mass[px][py] == 0 or mass[qx][qy]==0:
        return False
    
    hash1 = (px-1)*W + (py-1)
    hash2 = (qx-1)*W + (qy-1)
    return UF.same(hash1,hash2)



for i in range(Q):
    q = list(map(int,input().split()))
    if q[0]==1:
        r,c = q[1],q[2]
        query1(r,c)
        
    else: # q[0]==2
        ra,ca = q[1],q[2]
        rb,cb = q[3],q[4]
        answer = query2(ra,ca,rb,cb)
        if answer:
            print("Yes")
        else:
            print("No")
```

```python
from heapq import heappop, heappush

N, M = map(int,input().split())


E = [ [] for _ in range(N)]
for i in range(M):
    a,b,c = map(int,input().split())
#     0 index
    a -= 1
    b -= 1
    E[a].append((c,b))
    E[b].append((c,a))
    
def dijekstra(start):
#     点から点までのコスト
    dist = [10**18]*N
    dist[start] = 0
    HQ = []
    
#     (cost(start), start) を HQ に push
    heappush(HQ,(0, start))
    
    while HQ:
        
        _, now = heappop(HQ)
        for cost,idx in E[now]:
            if cost + dist[now] < dist[idx]:
                dist[idx] = dist[now] + cost
                
                heappush(HQ, (dist[idx],idx))
    
    return dist

D0 = dijekstra(0)
# dist[1] : 0~1までのコスト
DN = dijekstra(N-1)

# D0 + DN はだめ!
for d0, dn in zip(D0,DN):
    print(d0+dn)
                
```

# 復習

```python
N = int(input())
A = [iint(i) for i in input().split()]
Q = int(input())

A = sorted(A)
# comp = []
for i in range(Q):
    b = int(input())
    r = A[-1]
    l = A[0]
    m = (r+l)//2
   
    while r - l > 1:
        if m > b:
            l = m
        else:
            r = m
        m = (r+l)//2

    min_comp = min(abs(r-b),abs(l-b))
    print(min_comp)
        
```
